```python
from scipy.io import readsav
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df_1 = pd.read_csv('VERTICAL_PULITO.csv')
df_2 = pd.read_csv('HOR_PULITO.csv')
```

```python
#plotta il dataset pulito per ogni istante
def plot_clean(df_1,df_2,time,titolo):
    
    # Filtra i dati per il valore corrente di 'Time'
    df_1_block = df_1[df_1['Time'] == time]
        
    # Estrai X e Y per il blocco corrente
    X_block_V = df_1_block['PREL_VERT'].values
    Y_block_V = df_1_block['BRIGHT_VERT'].values

    # Filtra i dati per il valore corrente di 'Time'
    df_2_block = df_2[df_2['Time'] == time]
        
    # Estrai X e Y per il blocco corrente
    X_block_H = df_2_block['PREL_HOR'].values
    Y_block_H = df_2_block['BRIGHT_HOR'].values


    # Creazione della finestra con due sotto-grafici affiancati
    fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(14, 6))

    # Primo scatter plot: 'BRIGHT_VERT'
    ax1.scatter(X_block_V, Y_block_V, color='blue', label='Dati originali')
    #ax1.plot(X_V, y_pred_poly_3_V, color='red', label='Fit Polinomiale (grado 3)')
    ax1.set_xlabel('PREL Vert')
    ax1.set_ylabel('Brightness')
    ax1.set_title('Fit Polinomiale di Grado 3_Vert')
    ax1.legend(fontsize='small')
    ax1.grid(True)

    # Secondo scatter plot: 'BRIGHT_HOR'
    ax2.scatter(X_block_H, Y_block_H, color='blue', label='Dati originali')
    #ax2.plot(X_H, y_pred_poly_3_H, color='red', label='Fit Polinomiale (grado 3)')
    ax2.set_xlabel('PREL Hor')
    ax2.set_ylabel('Brightness')
    ax2.set_title('Fit Polinomiale di Grado 3_Hor')
    ax2.legend(fontsize='small')
    ax2.grid(True)


    # Aggiungi uno spazio tra i grafici se necessario
    plt.tight_layout()
    plt.suptitle(titolo, fontsize=16)
    plt.tight_layout()

    plt.show()
    
```

```python
for time in range(0, 100, 10):
  plot_clean(df_1,df_2,time,f'Dati Puiti Time={time}')
```
